```python
# Importing relevant frameworks and libraries
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
sns.set(rc={'figure.figsize':(12,6)})
pd.set_option('display.max_colwidth', None)
```

```python
def len_of_text(text_list):
  """
    Finds the number of words/tokens in the text
    input : list of strings/documents
    output : returns length of each string document
  """
  len_list = []
  for text in text_list : 
    temp_list = text.split()
    len_list.append(len(temp_list))
  return len_list

def plot_len_of_text(feature, lengths):
  """
    Plots the violinplot of the distribution of number of words in text feature 
    input : name of the feature and list of lengths of the documents/entries
    outpu : displays violin plot of distribution of length of documents
  """
  sns.violinplot(lengths)
  plt.xlabel(feature)
  plt.ylabel('Length / Number of words')
  plt.title('number of words in ' + feature)
  plt.show()

data = pd.read_csv('clean_data.csv')
data.dropna(inplace=True)
title_lengths = len_of_text(data['title'])
plot_len_of_text('title', title_lengths)
```

Observations :  

1. Majority of the job titles have two to eight words in them
2. The median number of words in job title is four. 
3. The inter-quartile range of number of words in job title is from three to 
six.
4. Their are some outliers that have around 12 to 14 words in them.  

```python
basic_lengths = len_of_text(data['basic_qualif'])
plot_len_of_text('basic qualifications', basic_lengths)
```

Observations : 

1. Majority of the basic job qualifications have less than 100 words. 
2. The median number of words in job description is 50. 
3. The inter-quartile range of number of words in job-description is from 40 to 75 words. 
4. There are very small number of recrods that have job description longer than 200. 


```python
pref_lengths = len_of_text(data['pref_qualif'])
plot_len_of_text('preferred qualifications', pref_lengths)
```

Observations :  

1. The majority of preferred qualifications have less than 180 words.
2. The median number of words in preferred qualifications is around 75. 
3. The inter quartile range for number of words in preferred qualifications is 50 to 90. 
4. There is a very small number of preferred qualifications where the number of words is more than equal to 200.  

```python
def display_wordCloud(feature, data):
  """Display word cloud
     Input : feature for which word cloud needs to be displayed and the dataframe
     Output : Displays the word cloud return None
  """
  text = ' '.join(data[feature].values)
  wordcloud = WordCloud().generate(text)
  # Display the generated image: take relative word frequencies into account, lower max_font_size
  wordcloud = WordCloud(background_color="white",max_words=len(text),max_font_size=40, relative_scaling=.5).generate(text)
  plt.figure()
  plt.title('word cloud for ' + feature)
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()

display_wordCloud('title', data)
```

```python
display_wordCloud('basic_qualif', data)
```

```python
display_wordCloud('pref_qualif', data)
```

Observations :  

1. Software engineer is the most common job title in our dataset 
2. Having a computer science background seems to be the most common basic qualification and it is for the software engineer role (as it the most common role found in the dataset). 
3. Having good communication skill, software development experience, distributed system and problem solving skills are some of the most preferred qualifications.
